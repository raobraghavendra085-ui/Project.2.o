```
import matplotlib.pyplot as plt
devices=['Mobile','laptop','desktop','table']
logins=[500,300,100,50]

plt.figure()
plt.pie(logins,labels=devices,autopct='%1.1f%%',startangle=90)
plt.title('online Education login Device Distribution')
plt.show()
```

```
days= ['Mon', 'Tus', 'wed','Thu', 'fri']
mobile_login = [100,120,130,110,140]

plt.figure()
plt.plot(days,mobile_login, marker='o')
plt.xlabel('Days')
plt.ylabel('Mobile Login')
plt.title('Mobile Login Trends')
plt.show()
```
